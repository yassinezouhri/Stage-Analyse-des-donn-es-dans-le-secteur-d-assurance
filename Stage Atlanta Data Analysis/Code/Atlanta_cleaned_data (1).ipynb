```python
import pdfplumber
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

with pdfplumber.open("C:/Users/HP/Downloads/Préz_Comités.pdf") as file:
    tables = []
    for i in file.pages:
        tables.extend(i.extract_tables())


dataframes = []
for table in tables:
    df = pd.DataFrame(table[1:], columns=table[0])
    dataframes.append(df)
```
